```python
def rgb_to_yuv(r, g, b):
    y = 0.299 * r + 0.587 * g + 0.114 * b
    u = 0.492 * (b - y)
    v = 0.877 * (r - y)
    return y, u, v

def yuv_to_rgb(y, u, v):
    r = y + 1.14 * v
    g = y - 0.395 * u - 0.581 * v
    b = y + 2.032 * u
    return r, g, b

# Example usage
rgb_values = (255, 0, 0)  # Example RGB values (Red)
yuv_values = rgb_to_yuv(*rgb_values)
print("RGB to YUV:", yuv_values)

converted_rgb_values = yuv_to_rgb(*yuv_values)
print("YUV to RGB:", converted_rgb_values)
```

```python
def serpentine(file_path):
    serpentine_bytes = []
    with open(file_path, 'rb') as file:
        # Read all bytes from the file
        image_bytes = file.read()
        
        # Read the bytes in a serpentine (zigzag) pattern
        width, height = 0, 0
        for byte in image_bytes:
            # Append the byte to the serpentine_bytes list
            serpentine_bytes.append(byte)
            
            # Check if the current position is at the end of the row
            if height % 2 == 0:
                width += 1
            else:
                width -= 1
            
            # Check if the current position is at the end of the image width
            if width >= 640:  # Assuming width of the image is 640 for example
                height += 1
                width -= 1
            elif width < 0:
                height += 1
                width += 1
        
    return bytes(serpentine_bytes)

# Example usage
input_image_path = '/Users/marina/Desktop/UNI/4rdyear/1rstTerm/CodAudioVideo/LABS/SISCOD/siscod_lab1.jpeg'
serpentine_image_bytes = serpentine(input_image_path)
print(serpentine_image_bytes)
```

```python
import subprocess

def compress_to_bw(input_file, output_file):
    # FFmpeg command to convert the image to grayscale with hard compression and save as JPEG
    ffmpeg_command = f'ffmpeg -i {input_file} -vf "format=gray" -q:v 1 {output_file}'
    
    # Run FFmpeg command
    subprocess.run(ffmpeg_command, shell=True)

# Example usage
input_image_path = '/Users/marina/Desktop/UNI/4rdyear/1rstTerm/CodAudioVideo/LABS/SISCOD/siscod_lab1.jpeg'
output_image_path = '/Users/marina/Desktop/UNI/4rdyear/1rstTerm/CodAudioVideo/LABS/SISCOD/output_bw_compressed1.jpg'

compress_to_bw(input_image_path, output_image_path)

```

```python
def run_length_encode(data):
    encoded_data = []
    count = 1
    for i in range(1, len(data)):
        if data[i] == data[i - 1]:
            count += 1
        else:
            encoded_data.extend([data[i - 1], count])
            count = 1
    encoded_data.extend([data[-1], count])
    return bytes(encoded_data)

# Example usage
input_bytes = b'\x01\x01\x01\x02\x03\x03\x04\x05\x05\x05'
encoded_bytes = run_length_encode(input_bytes)
print("Encoded bytes:", encoded_bytes)
```

data is the input series of bytes we want to encode, then the function iterates through the input data and counts consecutive occurrences of the same value, when a different value is encountered, it appends the previous value and its count to the encoded_data list.

Finally, it appends the last value and its count to the encoded_data list and returns the encoded bytes.

```python
import numpy as np
from PIL import Image
from scipy.fftpack import dct, idct

class DCTConverter:
    def __init__(self):
        pass
    
    def forward_dct(self, input_data):
        return dct(dct(input_data, axis=0, norm='ortho'), axis=1, norm='ortho')

    def inverse_dct(self, dct_data):
        return idct(idct(dct_data, axis=0, norm='ortho'), axis=1, norm='ortho')

# Load an image as a numpy array
image_path = '/Users/marina/Desktop/UNI/4rdyear/1rstTerm/CodAudioVideo/LABS/SISCOD/siscod_lab1.jpeg'
image = np.array(Image.open(image_path).convert('L'))  # 'L' converts the image to grayscale

# Create an instance of the DCTConverter class
dct_converter = DCTConverter()

# Forward DCT on the image
dct_image = dct_converter.forward_dct(image)

# Inverse DCT to reconstruct the image
reconstructed_image = dct_converter.inverse_dct(dct_image)

# Convert the reconstructed numpy array back to an image
reconstructed_image = Image.fromarray(np.uint8(reconstructed_image))

# Save the reconstructed image
reconstructed_image.save('reconstructed_image1.jpg')

```
